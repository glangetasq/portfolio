```python
class Test:
    
    def __init__(self, function):
        
        self.f = function
        
        
    def one_test(self, arg, expected_result):
        result = "GOOD" if (self.f(arg) == expected_result) else "WRONG"
        
        print(f"{result}: f({arg}) = {self.f(arg)} | expected: {expected_result}")
        
    
    def chunk_tests(self, args, expected_results):
        
        for arg, expected_result in zip(args, expected_results):
            self.one_test(arg, expected_result)
```

# 2. [Stock buy and sell](https://practice.geeksforgeeks.org/problems/stock-buy-and-sell-1587115621/1)

```python
def stock_buy_sell(A):

    buy_idx = 0
    buys_sells = []

    for i in range(1, len(A)):

        last_price = A[i-1]
        current_price = A[i]

        if current_price < last_price:
            buys_sells.append( (buy_idx, i-1) )
            buy_idx = i

        if buy_idx == i-1 and current_price == last_price:
            buy_idx = i


    if A[-1] > A[buy_idx]:
        buys_sells.append( (buy_idx, len(A)-1) )


    return [ (b, s) for b, s in buys_sells if b != s ]
```

```python
stockBuySell = Test(stock_buy_sell)

stockBuySell.one_test(
    [150, 120, 100, 180, 260, 310, 40, 535, 695], [(2, 5), (6, 8)]
)
```

# 3. [Total Decoding Messages](https://practice.geeksforgeeks.org/problems/total-decoding-messages1235/1)

```python
def count_ways(string):

    def count_ways_aux(string, memorization):
        
        if string not in memorization:
            
            a, b = int(string[0]), int(string[1])

            if a == 0:
                return -100000
            
            n_ways = 0


            if a <= 2 and b <= 6:
                n_ways += count_ways_aux(string[2:], memorization)
            if b > 0: # take into account that '10' and '20' can only be decoded 1 way
                n_ways += count_ways_aux(string[1:], memorization)

            memorization[string] = n_ways

        return memorization[string]


    memorization = { str(s):1 for s in range(1, 10) }
    memorization['10'] = memorization['20'] = memorization[''] = 1
    memorization['0'] = -1000000

    return max(count_ways_aux(string, memorization), 0)
```

```python
countWays = Test(count_ways)

countWays.one_test("123", 3)
countWays.one_test("27", 1)
countWays.one_test("0123", 0)
```

# 4. [Overlapping Rectangles](https://practice.geeksforgeeks.org/problems/overlapping-rectangles/0)

```python
class Rectangle:
    
    def __init__(self, xl, yl, xr, yr):
        
        self.xl, self.yl = xl, yl
        self.xr, self.yr = xr, yr
        
    
    def get_four_points(self):
        
        return [ (x, y) for x in (self.xl, self.xr) for y in (self.yl, self.yr) ]
    
    
    def is_point_in(self, x, y):
        
        xmin, ymin = min(self.xl, self.xr), min(self.yl, self.yr)
        xmax, ymax = max(self.xl, self.xr), max(self.yl, self.yr)
        
        return (xmin <= x <= xmax) and (ymin <= y <= ymax)

    
    @classmethod
    def from_line(cls, line):
        xl, yl, xr, yr = [ int(a) for a in line.split(' ') ]
        return cls(xl, yl, xr, yr)
        
```

```python
def _overlapping_rectangles(rec1, rec2):
    """
    Return 1 if any points of any rectangle is inside the other rectangle
    
    Note: if rec1 and rec2 share a corner, they'll be considered overalapping
    """
    
    
    test = False
    
    for x, y in rec1.get_four_points():
        test = test or rec2.is_point_in(x, y)
        
    for x, y in rec2.get_four_points():
        test = test or rec1.is_point_in(x, y)
        
    return 1 if test else 0

def overlapping_rectangles(lines):
    line1, line2 = lines
    rec1, rec2 = Rectangle.from_line(line1), Rectangle.from_line(line2)
    return _overlapping_rectangles(rec1, rec2)
```

```python
overlappingRectangles = Test(overlapping_rectangles)

overlappingRectangles.one_test(
    ["0 10 10 0", "5 5 15 0"], 1
)
overlappingRectangles.one_test(
    ["0 2 1 1", "-2 -3 -1 2"], 0
)
```

# 5. [Non Repeating Character](https://practice.geeksforgeeks.org/problems/non-repeating-character-1587115620/1)

```python
class UniqueCharacterStack:
    
    def __init__(self):
        
        self.stack = []
        self.already = set()
        
        
    def add(self, char):
        
        if char not in self.already:
            if char in self.stack:
                self.stack.remove(char)
                self.already.add(char)
            else:
                self.stack.append(char)
                
    def first_non_repeating(self):
        return self.stack[0] if self.stack else '$'
        
```

```python
def non_repeating_character(string):
    
    stack = UniqueCharacterStack()
    
    for c in string:
        stack.add(c)
        
    return stack.first_non_repeating()
```

```python
nonRepeatingCharacter = Test(non_repeating_character)

nonRepeatingCharacter.one_test('hello', 'h')
nonRepeatingCharacter.one_test('zxvczbtxyzvy', 'c')
nonRepeatingCharacter.one_test('zzzzzzzccdd', '$')
```

# 6. [Sum Tree](https://practice.geeksforgeeks.org/problems/sum-tree/1)

```python
class TreeNode:
    
    def __init__(self, data, left=None, right=None):
        
        self.data = data
        self.left, self.right = left, right
        
    def is_leaf(self):
        return self.left is None and self.right is None        
        
```

```python
def sum_tree(root):
    
    def recursive_sum_tree(root):
        if root.is_leaf():
            return True
        else:
            isSubSumTree = recursive_sum_tree(root.left) and recursive_sum_tree(root.right) 
            return isSubSumTree and (root.data == root.left.data + root.right.data)
        
    return 1 if recursive_sum_tree(root) else 0
```

# 7. [Flattening a Linked List](https://practice.geeksforgeeks.org/problems/flattening-a-linked-list/1)

```python
class Node:
    
    def __init__(self, data, bottom_=None, next_=None):
        self.data = data
        self.bottom = bottom_
        self.next = next_
        
    
```

```python
def build_linked_list(values):
    
    if not values:
        return Node(None)
    
    root = Node(values[0])
    current_node = root
    
    for data in values[1:]:
        current_node.bottom = Node(data)
        current_node = current_node.bottom
        
    return root
        
        
```

```python
def build_two_dimensional_linked_list(matrix):
    
    vertical_linked_list = [ build_linked_list(values) for values in matrix ]
    
    for i in range(len(vertical_linked_list)-1):
        vertical_linked_list[i].next = vertical_linked_list[i+1]
        
    return vertical_linked_list[0]
    
    
```

```python
def flatten(root):

    # If empty 2D-LL, then no flattening to do
    # If root.next is None, then root is already flattened
    if root is None or root.next is None:
        return root

    # Otherwise, root.next needs to be flatten first
    root.next = flatten(root.next)

    # Here, flattening a node with root.bottom AND root.next flattened
    
    # If root.bottom is None, then we just need to put the right side at the bottom, it'll be flattened
    if root.bottom is None:
        root.bottom = root.next
    # If bottom.data <= next.data then we move "down", and continue flattening
    elif root.bottom.data <= root.next.data:
        root.bottom.next = root.next
        root.next = None
        root.bottom = flatten(root.bottom)
    # If bottom.data > next.data, then we swap bottom with next and move "DOWN"
    else:
        root.next.next = root.bottom
        root.bottom = root.next
        root.next = None
        root.bottom = flatten(root.bottom)
            
    return root

```

```python
def print_ll(ll):
    c_node = ll
    values = []

    while c_node is not None:
        values.append(c_node.data)
        c_node = c_node.bottom
        
    print(' -> '.join([ str(x) for x in values ]))
```

```python
ll = build_two_dimensional_linked_list([
    [5, 7, 8, 30], 
    [10, 20],
    [19, 22, 50],
    [28, 35, 40, 45]
])

print_ll(flatten(ll))
```

# 8. [Binary Tree to Doubly Linked List](https://practice.geeksforgeeks.org/problems/intersection-point-in-y-shapped-linked-lists/1)

```python
class Node:
    
    def __init__(self, value, left=None, right=None):
        
        self.data = self.value = value
        self.left = left
        self.right = right
        
    
    def __str__(self):
        l = self.left.data if self.left else ' '
        r = self.right.data if self.right else ' '
        return "     {}    \n{}        {}".format(self.data, l, r)

        
        
```

```python
def print_ll(root):
    
    left_values, right_values = [], []
    current_node = root
    last_node = None
    
    while current_node:
        left_values.append(current_node.data)
        last_node = current_node
        current_node = current_node.right

        
    while last_node:
        right_values.append(last_node.data)
        last_node = last_node.left
        
        
    left_values = [ str(v) for v in left_values ]
    right_values = [ str(v) for v in right_values ]
    
    print(" -> ".join(left_values))
    print(" -> ".join(right_values))
```

```python
def bintree_to_dll(root, left_carry=None, right_carry=None):
    
    if root.right:
        root.right = bintree_to_dll(root.right, left_carry=root, right_carry=right_carry)
    elif right_carry: # root.right AND right_carry
        root.right = right_carry
        #right_carry.left = root
    
    if root.left:
        root.left = bintree_to_dll(root.left, left_carry=left_carry, right_carry=root)
        return root.left
    else:
        root.left = left_carry
        return root

    return root
    
    
```

```python
# Graph
graph = Node(10)
graph.left = Node(12)
graph.left.left = Node(25)
graph.left.right = Node(30)
graph.right = Node(15)
graph.right.left = Node(36)
graph.right.left.left = Node(-1)
graph.right.left.right = Node(100)

dll = bintree_to_dll(graph)
```

```python
from collections import deque

def build_tree(s):
    
    if (len(s) == 0 or s[0] == 'N'):
        return None
    
    ip = list(map(str, s.split()))
    
    root = Node(int(ip[0]))
    size = 0
    q = deque()
    
    q.append(root)
    size += 1
    
    i = 1
    while (size > 0 and i < len(ip)):
        
        currNode = q[0]
        q.popleft()
        size -= 1
        
        currVal = ip[i]
        
        if currVal != 'N':
            
            currNode.left = Node(int(currVal))
            
            q.append(currNode.left)
            size += 1
            
        i += 1
        if i >= len(ip):
            break
            
        currVal = ip[i]
        
        if currVal != 'N':
            
            currNode.right = Node(int(currVal))
            
            q.append(currNode.right)
            size += 1
        
        i += 1
        
    return root
```

# 9. [Minimum element from stack](https://practice.geeksforgeeks.org/problems/get-minimum-element-from-stack/1/?difficulty[]=1&page=1&company[]=Goldman%20Sachs&query=difficulty[]1page1company[]Goldman%20Sachs)

Implement push, pop, get_min methods in O(1)

Here is a solution with O(1) time complexity, but O(n) extra-space complexity.
GeeksForGeeks have another method in O(1) in time AND extra-space complexity.

```python
class MinElementStack:
    
    def __init__(self):
        
        self.stack = []
        self.min_elements = []
        self.current_min = None
        
    
    def pop(self):
        self.min_elements.pop()
        return self.stack.pop()
    
    
    def push(self, value):
        
        self.current_min = min(value, self.current_min) if self.current_min else value
        self.min_elements.append(self.current_min)
        self.stack.append(value)
        
        
    def get_min(self):
        return self.current_min
        
        
```

```python
minstack = MinElementStack()

minstack.push(2)
print(minstack.get_min())
minstack.push(3)
print(minstack.get_min())
minstack.pop()
print(minstack.get_min())
minstack.push(1)
print(minstack.get_min())
```

# 10. [Knight Walk](https://practice.geeksforgeeks.org/problems/knight-walk4521/1/?difficulty[]=1&page=1&company[]=Goldman%20Sachs&query=difficulty[]1page1company[]Goldman%20Sachs#)

```python
class Solution:
    
    BEHAVIORAL_THRESHOLD = 5
    
    def norm1(self, posA, posB):
        
        xA, yA = posA
        xB, yB = posB
        
        return max(abs(xB-xA), abs(yB-yA))
        
    
    def next_positions(self, KnightPos, N):
        
        incr = [-2, -1, 1, 2]
        
        x, y = KnightPos
        
        positions = []
        
        for x_incr in incr:
            for y_incr in incr:
                
                isPossible = (abs(x_incr) != abs(y_incr))
                isPossible = isPossible and (0 < x + x_incr <= N)
                isPossible = isPossible and (0 < y + y_incr <= N)
                
                if isPossible:
                    positions.append( (x+x_incr, y+y_incr) )
                    
        return positions
        
        
        
    def minStepToReachTargetUtil(self, KnightPos, TargetPos, N, n_steps=0, visited=[], memorization={}):
        
        if KnightPos not in memorization:
            
            NextKnightPositions = self.next_positions(KnightPos, N)
            
            
            # When knight is to far, we can consider position that get closer to target only w.l.o.g
            current_norm = self.norm1(KnightPos, TargetPos)
            
            print(KnightPos, current_norm, NextKnightPositions)
            if current_norm > self.BEHAVIORAL_THRESHOLD:
                NextKnightPositions = [ 
                    pos for pos in NextKnightPositions if self.norm1(pos, TargetPos) < current_norm
                ]
            
            # Get n_steps for the considered move
            all_steps = []
            for NextKnightPos in NextKnightPositions:
                all_steps.append( self.minStepToReachTargetUtil(
                        NextKnightPos, TargetPos, N, n_steps+1, memorization
                ))
                
            print(KnightPos, NextKnightPositions)
                
            memorization[KnightPos] = 1 + min(all_steps)
        
        return memorization[KnightPos]
                
                
            
        
    def minStepToReachTarget(self, KnightPos, TargetPos, N):

        KnightPos = tuple(KnightPos)
        TargetPos = tuple(TargetPos)


        return self.minStepToReachTargetUtil(
            KnightPos, TargetPos, N, n_steps=0, memorization={TargetPos:0}
           )
```

# 11. [Check for BST](https://practice.geeksforgeeks.org/problems/check-for-bst/1/?difficulty[]=1&page=1&company[]=Goldman%20Sachs&query=difficulty[]1page1company[]Goldman%20Sachs)

```python
class Node:
    
    def __init__(self, data, left=None, right=None):
        
        self.left = left
        self.data = data
        self.right = right
        
        
```

```python
def is_bst(root, minimum=None, maximum=None):
    
    if root is None:
        return True
    
    
    isNodeBST = True
    
    if minimum:
        isNodeBST = isNodeBST and (root.data > minimum)
    if maximum:
        isNodeBST = isNodeBST and (root.data < maximum)
        
    isLeftBST = is_bst(root.left, minimum=minimum, maximum=root.data)
    isRightBST = is_bst(root.right, minimum=root.data, maximum=maximum)
        
    
    return isNodeBST and isLeftBST and isRightBST
    
    
```

```python
rootA = Node(2)
rootA.left = Node(1)
rootA.right = Node(3)

rootB = Node(2)
rootB.right = Node(3)
rootB.right.right = Node(5)
rootB.right.left = Node(4)

isBST = Test(is_bst)

isBST.one_test(rootA, True)
isBST.one_test(rootB, False)
```

# 12. [Egg Dropping Puzzle](https://practice.geeksforgeeks.org/problems/egg-dropping-puzzle-1587115620/1/?difficulty[]=1&page=1&company[]=Goldman%20Sachs&query=difficulty[]1page1company[]Goldman%20Sachs)

```python
def egg_drop(prm):
    
    n, k = prm

    memorization = [ (k+1)*[None] for _ in range(n+1) ]
    
    # Initialization
    
    # 0 floor, n eggs => 0 attemps
    for i in range(n+1):
        memorization[i][0] = 0

    # 1 egg, k floors => k attempts on worse case
    for j in range(k+1):
        memorization[1][j] = j

    # Recursion
    
    for i in range(2, n+1):
        for j in range(1, k+1):
            
            min_step = k + 1 # MAX 
            
            for step in range(1, j+1):
                
                try:
                    worse_case = max(
                        memorization[i-1][step-1], # if egg breaks, only step-1 floors to check
                        memorization[i][j-step], # if egg doesn't break, has n_floors-step floors to check
                    )

                    min_step = min(min_step, worse_case)
                except:
                    print(i, j, step, j-step, memorization[i])
            
            memorization[i][j] = 1 + min_step
    
    
    return memorization[-1][-1]
```

```python
eggDrop = Test(egg_drop)

eggDrop.one_test((2,10), 4)
eggDrop.one_test((3,5), 3)
```
